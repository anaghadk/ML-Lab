```python
import pandas as pd
from sklearn.preprocessing import OrdinalEncoder
df=pd.read_csv('Dataset of Diabetes .csv')
print("#1 To display information of all columns")
print(df.info())
print("#2 To display statistical information of all numerical")
print(df.describe())
print("#3 To display the count of unique labels for “Gender” column")
print(df["Gender"].value_counts())
print("#4 ")
missing_values = df.isnull().sum()
print(missing_values[missing_values > 0])
#print(df.isnull().sum()[df.isnull().sum()>0])
print("#5 Null values in each column")
print(df.isnull().sum())
print("#6")
df["AGE"].fillna(df["AGE"].mean())
print(df.head())

print("#7 Which columns in the dataset had missing values? How did you handle them ?")
ddf = df.select_dtypes(include=['category', 'object']).columns.tolist()
print(ddf)

print("#8")


# Initialize encoder
ordinal_encoder = OrdinalEncoder()

# Apply encoding
df[ddf] = ordinal_encoder.fit_transform(df[ddf])
print(df.head())


```
