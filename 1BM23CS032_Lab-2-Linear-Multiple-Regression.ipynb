```python
import pandas as pd
import matplotlib.pyplot as plt
#from scikitlearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression

df=pd.read_csv('canada_per_capita_income.c sv')
df.head()
```

```python
# Prepare data
X = df[['year']]
y = df['per capita income (US$)']

# Train model
model = LinearRegression()
model.fit(X, y)

predicted_income = model.predict([[2020]])
print(predicted_income)

# Predict for plotting regression line
y_pred = model.predict(X)

# Plot
plt.figure(figsize=(8,5))
plt.scatter(X, y, color='blue', label='Actual Data')
plt.plot(X, y_pred, color='red', linewidth=2, label='Regression Line')

plt.title("Canada Per Capita Income Prediction")
plt.xlabel("Year")
plt.ylabel("Per Capita Income (US$)")
plt.legend()
plt.grid(True)
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
df=pd.read_csv('salary.csv')
df.head()
```

```python
df.isnull().sum()
df['YearsExperience'].fillna(df['YearsExperience'].mean(),inplace=True)
df.isnull().sum()
```

```python
X = df[['YearsExperience']]
y = df['Salary']

model_sal = LinearRegression()
model_sal.fit(X, y)

predicted = model_sal.predict([[12]])
print('Salary predicted for 12 years of experience')
print(predicted)

# Create smooth regression line
import numpy as np
X_range = np.linspace(X.min(), X.max(), 100).reshape(-1,1)
y_pred_range = model_sal.predict(X_range)

# Plot
plt.figure(figsize=(8,5))
plt.scatter(X, y, color='green', label='Actual Salary Data')
plt.plot(X_range, y_pred_range, color='red', linewidth=2, label='Linear Regression Line')

plt.title("Salary vs Years of Experience")
plt.xlabel("Years of Experience")
plt.ylabel("Salary")
plt.legend()
plt.grid(True)
plt.show()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Load dataset
df = pd.read_csv('salary.csv')
df['YearsExperience'].fillna(df['YearsExperience'].mean(),inplace=True)
# Extract variables
X = df['YearsExperience'].values
y = df['Salary'].values

# Calculate means
x_mean = np.mean(X)
y_mean = np.mean(y)

# Calculate slope (b1)
numerator = np.sum((X - x_mean) * (y - y_mean))
denominator = np.sum((X - x_mean) ** 2)
b1 = numerator / denominator

# Calculate intercept (b0)
b0 = y_mean - b1 * x_mean

print("Slope (b1):", b1)
print("Intercept (b0):", b0)

# Predict salary for 12 years
predicted = b0 + b1 * 12
print("Salary predicted for 12 years of experience:")
print(predicted)

# Create smooth regression line
X_range = np.linspace(min(X), max(X), 100)
y_pred_range = b0 + b1 * X_range

# Plot
plt.figure(figsize=(8,5))
plt.scatter(X, y, color='green', label='Actual Salary Data')
plt.plot(X_range, y_pred_range, color='red', linewidth=2, label='Regression Line')

plt.title("Salary vs Years of Experience")
plt.xlabel("Years of Experience")
plt.ylabel("Salary")
plt.legend()
plt.grid(True)
plt.show()
```

```python
import pandas as pd
df1=pd.read_csv('hiring.csv')
df1.head()
```

```python
df1.isnull().sum()
```

```python
df1['experience'].fillna('zero',inplace=True)
df1['test_score(out of 10)'].fillna(df1['test_score(out of 10)'].mean(),inplace=True)
df1.isnull().sum()
```

```python
import numpy as np
from sklearn.linear_model import LinearRegression
word_to_num = {
    'zero': 0,
    'one': 1,
    'two': 2,
    'three': 3,
    'four': 4,
    'five': 5,
    'six': 6,
    'seven': 7,
    'eight': 8,
    'nine': 9,
    'ten': 10,
    'eleven': 11,
    'twelve': 12
}

df1['experience'] = df1['experience'].map(word_to_num)
X = df1[['experience', 'test_score(out of 10)', 'interview_score(out of 10)']]
y = df1['salary($)']

# -------------------------------
# Model Training
# -------------------------------

model = LinearRegression()
model.fit(X, y)

# -------------------------------
# Predictions
# -------------------------------

pred1 = model.predict([[2, 9, 6]])
pred2 = model.predict([[12, 10, 10]])

print("Salary prediction for 2yr exp, 9 test, 6 interview:")
print(pred1)

print("Salary prediction for 12yr exp, 10 test, 10 interview:")
print(pred2)
```
